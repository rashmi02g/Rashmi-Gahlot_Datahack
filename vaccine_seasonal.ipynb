```python
df=pd.read_csv('training_set_features.csv')
df2=pd.read_csv('training_set_labels.csv')
```

```python
df=df.apply(lambda x: x.fillna(x.mode()[0]), axis=0)
```

```python
X_seasonal=df.drop(columns=['employment_industry','employment_occupation','opinion_xyz_vacc_effective','opinion_xyz_risk','opinion_xyz_sick_from_vacc'])
```

```python
columns_to_encode = ['age_group','education','sex','income_poverty','employment_status','census_msa','race','rent_or_own','marital_status','hhs_geo_region']
```

```python
df_encoded = pd.get_dummies(X_seasonal, columns=columns_to_encode).astype(int)
```

```python
original_columns = list(X_seasonal.columns)
original_encoded_columns = [col for col in df_encoded.columns if col not in original_columns]
df_with_names = X_seasonal[original_columns].join(df_encoded[original_encoded_columns])
display(df_with_names.head())
```

```python
new_df=df_with_names.drop(columns=['age_group','education','sex','income_poverty','employment_status','census_msa','race','rent_or_own','marital_status','hhs_geo_region'])
```

```python
new_df.shape
```

```python
new_df.head()
```

```python
#x=new_df.drop(columns=['seasonal_vaccine'])
#y1=new_df.loc[:,'seasonal_vaccine']
y=df2.loc[:,'seasonal_vaccine']
```

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
clf = BernoulliNB()
clf.fit(new_df,y)
```

```python
y1_pred = clf.predict(new_df)
```

```python
accuracy = accuracy_score(y, y1_pred)
```

```python
accuracy
```

```python
probabilities = clf.predict_proba(new_df)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model1 = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model1.fit(new_df,y)
feature_importance = pd.DataFrame({
    'Feature': new_df.columns,
    'Importance': rf_model1.feature_importances_
})
feature_importance = feature_importance.sort_values(by='Importance', ascending=False)
print(feature_importance)
```
